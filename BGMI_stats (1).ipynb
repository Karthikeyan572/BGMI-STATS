```python
from sklearn.preprocessing import PowerTransformer

column = ['Matches_Played','Top_10s','Revives','Wins' ,'Distance_Traveled','Weapons_Used','Time_Survived' ]

pt = PowerTransformer(method='yeo-johnson')

for col in column:
    df[col + '_yeojhonson'] = pt.fit_transform(df[[col]])

df.head()
```

```python
df['kd_ratio'] = df['Kills']/df['Deaths']
df.head()
```

```python
df['wins_prob'] = round((df['Wins']/(df['Matches_Played']) *100), 2)
df.head()
```

```python
df['Losses'] = df['Matches_Played'] - df['Wins']
df.head()
```

```python
df['loss_prob'] = round((df['Losses']/(df['Matches_Played']) *100), 2)
df.head()
```

```python
cols = df.columns.tolist()
cols
```

```python
cols.remove('Losses')
cols.insert(8,'Losses')
df = df[cols]
```

```python
cols.remove('kd_ratio')
cols.insert(4,'kd_ratio')
df = df[cols]
```

```python
cols.remove('wins_prob')
cols.insert(9,'wins_prob')
df = df[cols]
```

```python
cols.remove('loss_prob')
cols.insert(11,'loss_prob')
df = df[cols]
```

```python
df.head()
```

```python
df['Score'] = ( (df['Wins'] * 0.5) - (df['Losses'] * 0.5) + (df['Kills'] * 0.4) + (df['Damage_Dealt']*0.3) + (df['Time_Survived']*0.3) + (df['Top_10s'] * 0.2) + (df['Revives'] * 0.1) ).astype(int)
df.head()
```

```python
df.sort_values(by='Score', ascending=False, inplace=True)
print("Top 5 players")
display(df.head(5))
```

```python
df['Score'].describe()
```

```python
plt.figure(figsize=(8,6))
plt.bar(df['Player_Name'][:5],df['kd_ratio'][:5])
plt.title('KD Ratio of players')
plt.xlabel('Player Name')
plt.ylabel('KD Ratio')
plt.xticks(rotation=90)
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.regplot(x='Kills', y='Wins', data=df, ax=axes[0], color='red', scatter_kws={'color': 'blue'})
axes[0].set_title('Wins vs Kills')
axes[0].set_xlabel('Kills')
axes[0].set_ylabel('Wins')
axes[0].grid(True)

sns.regplot(x='Kills', y='Damage_Dealt', data=df, ax=axes[1], color='purple', scatter_kws={'color': 'green'})
axes[1].set_title('Damage Dealt vs Kills')
axes[1].set_xlabel('Kills')
axes[1].set_ylabel('Damage Dealt')
axes[1].grid(True)

sns.regplot(x='Kills', y='Headshots', data=df, ax=axes[2], color='orange', scatter_kws={'color': 'brown'})
axes[2].set_title('Headshots vs Kills')
axes[2].set_xlabel('Kills')
axes[2].set_ylabel('Headshots')
axes[2].grid(True)

plt.show()
```

```python
numeric_cols = ['Wins', 'Kills', 'Top_10s', 'Matches_Played', 'Damage_Dealt', 'Weapons_Used',	'Time_Survived',  'Headshots']
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation of Player Statistics')
plt.show()
```

```python
# Save plotted figures to a 'plots' folder so they can be embedded in README.md
import os
plots_dir = os.path.join(os.getcwd(), 'plots')
os.makedirs(plots_dir, exist_ok=True)

# Recreate and save the histograms for the numeric columns
plt.figure(figsize=(16, 12))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(4, 2, i)
    sns.histplot(df[col].dropna(), kde=True, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
hist_path = os.path.join(plots_dir, 'histograms.png')
plt.tight_layout()
plt.savefig(hist_path, bbox_inches='tight', dpi=150)
print(f'Saved: {hist_path}')
plt.close()

# Recreate and save the boxplots
plt.figure(figsize=(16, 10))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(4, 2, i)
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
box_path = os.path.join(plots_dir, 'boxplots.png')
plt.tight_layout()
plt.savefig(box_path, bbox_inches='tight', dpi=150)
print(f'Saved: {box_path}')
plt.close()

# Save KD ratio bar for top 5
try:
    fig, ax = plt.subplots(figsize=(8,6))
    ax.bar(df['Player_Name'][:5], df['kd_ratio'][:5])
    ax.set_title('KD Ratio of players')
    ax.set_xlabel('Player Name')
    ax.set_ylabel('KD Ratio')
    plt.xticks(rotation=90)
    kd_path = os.path.join(plots_dir, 'kd_ratio_top5.png')
    fig.savefig(kd_path, bbox_inches='tight', dpi=150)
    print(f'Saved: {kd_path}')
    plt.close(fig)
except Exception as e:
    print('Could not save KD ratio plot:', e)

# Save correlation heatmap
try:
    corr_matrix = df[numeric_cols].corr()
    fig, ax = plt.subplots(figsize=(8,6))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, ax=ax)
    ax.set_title('Correlation of Player Statistics')
    corr_path = os.path.join(plots_dir, 'correlation_heatmap.png')
    fig.savefig(corr_path, bbox_inches='tight', dpi=150)
    print(f'Saved: {corr_path}')
    plt.close(fig)
except Exception as e:
    print('Could not save heatmap:', e)
```
