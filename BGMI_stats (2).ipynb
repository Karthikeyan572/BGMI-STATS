```python
from sklearn.preprocessing import PowerTransformer

column = ['Matches_Played',	'Kills',	'Deaths',	'Assists',	'Damage_Dealt',	'Headshots',	'Wins',	'Top_10s',	'Revives',	'Distance_Traveled',	'Weapons_Used',	'Time_Survived'	]

pt = PowerTransformer(method='yeo-johnson')

for col in column:
    df[col + '_yeojhonson'] = pt.fit_transform(df[[col]])

df.head()
```

```python
column = ['Matches_Played_yeojhonson',	'Kills_yeojhonson',	'Deaths_yeojhonson',	'Assists_yeojhonson',	'Damage_Dealt_yeojhonson',	'Headshots_yeojhonson',	'Wins_yeojhonson',	'Top_10s_yeojhonson',	'Revives_yeojhonson',	'Distance_Traveled_yeojhonson',	'Weapons_Used_yeojhonson',	'Time_Survived_yeojhonson'	]
skew_values = df[column].skew()
print(skew_values)
```

```python
plt.figure(figsize=(8,6))

sns.histplot(df['Matches_Played_yeojhonson'], kde=True, color='skyblue')
plt.title('After Yeo-Johnson Transformation: Matches_Played')
plt.xlabel('Matches_Played')
plt.ylabel('Frequency')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Revives_yeojhonson'], kde=True, color='skyblue')
plt.title('Distribution of Revives After Yeo-Johnson Transformation')
plt.xlabel('Revives (Yeo-Johnson Transformed)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Wins_yeojhonson'], kde=True, color='skyblue')
plt.title('Distribution of wins sfter Yeo-Johnson Transformation')
plt.xlabel('wins (Yeo-Johnson Transformed)')
plt.ylabel('Frequency')
plt.show()
```

```python
df['kd_ratio'] = df['Kills']/df['Deaths']
df.head()
```

```python
df['wins_prob'] = round((df['Wins']/(df['Matches_Played']) *100), 2)
df.head()
```

```python
df['Losses'] = df['Matches_Played'] - df['Wins']
df.head()
```

```python
df['loss_prob'] = round((df['Losses']/(df['Matches_Played']) *100), 2)
df.head()
```

```python
cols = df.columns.tolist()
cols
```

```python
cols.remove('Losses')
cols.insert(8,'Losses')
df = df[cols]
```

```python
cols.remove('kd_ratio')
cols.insert(4,'kd_ratio')
df = df[cols]
```

```python
cols.remove('wins_prob')
cols.insert(9,'wins_prob')
df = df[cols]
```

```python
cols.remove('loss_prob')
cols.insert(11,'loss_prob')
df = df[cols]
```

```python
df.head()
```

```python
df['Score'] = ( (df['Wins'] * 0.5) - (df['Losses'] * 0.5) + (df['Kills'] * 0.4) + (df['Damage_Dealt']*0.3) + (df['Time_Survived']*0.3) + (df['Top_10s'] * 0.2) + (df['Revives'] * 0.1) ).astype(int)
df.head()
```

```python
df.sort_values(by='Score', ascending=False, inplace=True)
print("Top 5 players")
display(df.head(5))
```

```python
plt.figure(figsize=(8,6))
plt.bar(df['Player_Name'][:5],df['kd_ratio'][:5])
plt.title('KD Ratio of players')
plt.xlabel('Player Name')
plt.ylabel('KD Ratio')
plt.xticks(rotation=90)
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.regplot(x='Kills', y='Wins', data=df, ax=axes[0], color='red', scatter_kws={'color': 'blue'})
axes[0].set_title('Wins vs Kills')
axes[0].set_xlabel('Kills')
axes[0].set_ylabel('Wins')
axes[0].grid(True)

sns.regplot(x='Kills', y='Damage_Dealt', data=df, ax=axes[1], color='purple', scatter_kws={'color': 'green'})
axes[1].set_title('Damage Dealt vs Kills')
axes[1].set_xlabel('Kills')
axes[1].set_ylabel('Damage Dealt')
axes[1].grid(True)

sns.regplot(x='Kills', y='Headshots', data=df, ax=axes[2], color='orange', scatter_kws={'color': 'brown'})
axes[2].set_title('Headshots vs Kills')
axes[2].set_xlabel('Kills')
axes[2].set_ylabel('Headshots')
axes[2].grid(True)

plt.show()
```

```python
numeric_cols = ['Wins', 'Kills', 'Top_10s', 'Matches_Played', 'Damage_Dealt', 'Weapons_Used',	'Time_Survived',  'Headshots']
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation of Player Statistics')
plt.show()
```
